```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv('wine.csv')
data
```

```python
data1=data.iloc[:,1:]
data1.head()
```

```python
# Converting to arrays
data_array=data1.values
```

```python
data_array
```

```python
# Normalizing the data
data_normal=scale(data_array)
data_normal
```

```python
# Taking PCA values
pca=PCA(n_components=13)
pca_values=pca.fit_transform(data_normal)
pca_values
```

```python
# Calcualting Variance 
var=pca.explained_variance_ratio_
var
```

```python
# Calculating Cumulative Variance
cumvar=np.cumsum(np.round(var,4)*100)
cumvar
```

```python
# Final database
data_final=pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pca1','pca2','pca3']),data[['Type']]],axis=1)
data_final
```

# Hierarchial Clustering

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(data_normal,'complete','euclidean')
plt.figure(figsize=(50,8))
sch.dendrogram(z)
plt.show()
```

## Inference

#Hierarchial Clustering is not possible in this case since we will have numerous clusters

# K-Means Clustering

```python
from sklearn.cluster import KMeans
WCSS=[]
for i in range (1,11):
    clf=KMeans(n_clusters=i)
    clf.fit(data_normal)
    WCSS.append(clf.inertia_)
plt.figure(figsize=(20,8))
plt.plot(range(1,11),WCSS)
```

```python
# Taking no. of clusters=3
clf=KMeans(n_clusters=3)
clf.fit_transform(data_normal)
```

```python
clf.labels_
```

```python
# Final table with clustered values
data['clust']=clf.labels_
data
```
